```python
1: In "oops", a class is a blueprint or template  for creating object.
   An object is an instance of a class and can contain
    data and specific functions to the class.
     here 'Classes' and 'objects' are the two main components of OOP.
        
   A class defines the properties and behaviors of objects.
   A objects are instances of a class that contain real data.
    
example:
    
class Car:
    def __init__(self, make, model, year):
        self.make = make
        self.model = model
        self.year = year
        
    def start(self):
        print("Starting the car")
        
    def stop(self):
        print("Stopping the car")    
    
```

```python
bmw = Car("BMW", "3 Series", 2019)
print("Make:", bmw.make)
print("Model:", bmw.model)
print("Year:", bmw.year)
bmw.start()
bmw.stop()
```

```python
audi = Car("Audi", "A4", 2020)
print("Make:", audi.make)
print("Model:", audi.model)
print("Year:", audi.year)
audi.start()
audi.stop()
```

```python
output:
    
>Make: BMW
Model: 3 Series
Year: 2019
Starting the car
Stopping the car

>Make: Audi
Model: A4
Year: 2020
Starting the car
Stopping the car
```

```python
2: The four pillars of Object-Oriented Programming system (OOPS) 
  1. "Abstraction"
        Hides complex implementation details from the user 
        and only exposes the essential information...
  2. "Encapsulation"
       Encapsulates the data and behavior within an object,
    providing a secure way to access and modify the data...
  3. "Inheritance"
       Allows the properties and methods of a parent class to be passed down
    to child classes,reducing code duplication and improving code reuse...
  4."Polymorphism"  
      Enables objects to take on multiple forms, such as
    overriding or operator overloading methods...
      This allows objects to be treated as objects of the parent class,
        even though they are actually objects of a child class.
    
```

```python
3: >The __init__() function is used in Python as a constructor method 
    to initialize the state of an 'object'.
   >It is automatically called when an object is created from a class
    and it allows the class to initialize the attributes of the object.
   > The __init__() function is used to set default values for the attributes 
     of the object, which can then be modified later if necessary.
    
    
 The function takes the self argument as its first parameter,
    which refers to the object being created...
    
    Example:

    
    class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age

  
Here, the __init__() function takes two arguments name and age, and
  sets them as the initial values of the 'object's name and age attributes...
    
    
  "person = Person("John Doe", 30)"
  
 Here, the person object is created and the __init__() function is automatically
called to set the initial values of the object's name and age attributes.   
    
    
    
    
```

```python
4: >self is used in  "OOPS" to refer to the instance of the class
   that is being executed.
   >In Python, "self" is used as the first parameter in every method 
defined in a class, and it refers to the instance of the class
   for which the method was called. 
   >Using "self" allows the method to access and modify the attributes
    and behavior of the object for which it was called...
    
Example:
    
 class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age
    
    def say_hello(self):
        print("Hello, my name is {} and I am {} years old.".format(self.name, self.age))

        
        
person = Person("prathamesh", 23)
person.say_hello()

output:
    
  Hello, my name is prathamesh and I am 23 years old.
        
    
    
    
```

```python
5: >Inheritance is a mechanism in object-oriented programming,
where an object inherits the properties and behavior of its parent object...
   >It allows a new class to be derived from an existing class by inheriting
    all its attributes and behaviors, thereby creating a hierarchy of classes
    that can be easily maintained and managed...
```

```python
1."Single InheritanceIn" 
    Single inheritance, a subclass inherits the properties and
    behavior of a single parent class.
 
Example:
    
class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age

class Employee(Person):
    def __init__(self, name, age, salary):
        Person.__init__(self, name, age)
        self.salary = salary

emp = Employee("Prathamesh", 23, 50000)
print(emp.name)
print(emp.age)
print(emp.salary)
"outout":
Prathamesh
23
50000

In this example, the Employee class inherits the properties (name and age)
from the Person class, and it adds its own property salary.
```

```python
2."Multi-level Inheritance"
    Multi-level inheritance is a type of inheritance where a derived class is
    created from a derived class of another class.
      In this type of inheritance, one class inherits the properties and
    behavior from its parent class and then another class inherits 
    the properties from the derived class...
    
    
Example:    
    
class Animal:
    def __init__(self, name):
        self.name = name
        
    def show_name(self):
        print(f"Name: {self.name}")
        
class Mammal(Animal):
    def __init__(self, name, type_of_mammal):
        Animal.__init__(self, name)
        self.type_of_mammal = type_of_mammal
        
    def show_type(self):
        print(f"Type of mammal: {self.type_of_mammal}")
        
class Dog(Mammal):
    def __init__(self, name, breed):
        Mammal.__init__(self, name, "Canine")
        self.breed = breed
        
    def show_breed(self):
        print(f"Breed: {self.breed}")
        
dog = Dog("Bholu", "Labrador")
dog.show_name()
dog.show_type()
dog.show_breed()

"output":
    
Name: Bholu
Type of mammal: Canine
Breed: Labrador   


In this example, the Dog class inherits from the Mammal class,
which in turn inherits from the Animal class.
This forms a chain of inheritance where the properties and
  methods of the parent classes can be accessed in the child class.
```

```python
3."Hierarchical Inheritance"
   In hierarchical inheritance, a single parent class is inherited
    by multiple subclasses...
    
Example:
class Animal:
    def __init__(self, name, species):
        self.name = name
        self.species = species

    def show_animal_details(self):
        print("Name:", self.name)
        print("Species:", self.species)

class Dog(Animal):
    def __init__(self, breed, name, species):
        Animal.__init__(self, name, species)
        self.breed = breed

    def show_dog_details(self):
        print("Breed:", self.breed)

class Labrador(Dog):
    def __init__(self, color, breed, name, species):
        Dog.__init__(self, breed, name, species)
        self.color = color

    def show_labrador_details(self):
        print("Color:", self.color)

dog = Labrador("white", "Labrador", "Bholu", "Mammal")
dog.show_animal_details()
dog.show_dog_details()
dog.show_labrador_details()


"Output":
Species: Mammal
Breed: Labrador
Color: white    
```

```python
4."Multiple inheritance"
    In multiple inheritance, a class can inherit properties and methods from
    more than one parent class...

    
Example: vehicle, engine , car

class Vehicle:
    def __init__(self, name, wheels):
        self.name = name
        self.wheels = wheels

    def display(self):
        print("Name:", self.name)
        print("Wheels:", self.wheels)

class Engine:
    def __init__(self, power, capacity):
        self.power = power
        self.capacity = capacity

    def display(self):
        print("Power:", self.power)
        print("Capacity:", self.capacity)

class Car(Vehicle, Engine):
    def __init__(self, name, wheels, power, capacity):
        Vehicle.__init__(self, name, wheels)
        Engine.__init__(self, power, capacity)

c = Car("Maruti", 4, 90, 1000)
c.display()

"Output":
    
 Name: Maruti
Wheels: 4   

In the above example, the class "Car" is inheriting from both the classes
"Vehicle" and "Engine", and it can access the properties and methods of 
both the parent classes...


```

```python
5. "Hybrid inheritance"
    Hybrid inheritance is a combination of multiple inheritance and
    hierarchical inheritance.
In hybrid inheritance, a single class inherits from multiple classes, and
at the same time, the inherited classes also inherit from other classes. 

Example:

    
class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age

class Teacher(Person):
    def __init__(self, name, age, subject):
        Person.__init__(self, name, age)
        self.subject = subject

class Engineer(Person):
    def __init__(self, name, age, company):
        Person.__init__(self, name, age)
        self.company = company

class EngineerTeacher(Teacher, Engineer):
    def __init__(self, name, age, subject, company):
        Teacher.__init__(self, name, age, subject)
        Engineer.__init__(self, name, age, company)

        
>In this example, we have three classes: Person, Teacher, and Engineer.
The Teacher class inherits from Person, and the Engineer class also inherits
from Person. The EngineerTeacher class inherits from both Teacher and
Engineer.
    
    
    
    
    
    
```
